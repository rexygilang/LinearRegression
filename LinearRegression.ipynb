```python
import pandas as pd

df = pd.read_csv("Latihan01.csv", decimal = ",") #load dataset
df.head() #menampilkan 5 data teratas
```

```python
df.tail() #menampilkan 5 data terbawah
```

```python
ah = df.values #menjadikan nilai array
ah
```

```python
ah+1
```

```python
#membagi data

batas = ah.shape[0]*8//10

x = ah[:,:-1]
y = ah[:,-1]

x_train = x[ :batas]
x_test = x[batas: ]
y_train = y[ :batas]
y_test = y[batas: ]

x_train.shape
```

```python
y_train.shape
```

```python
#membuat model
from sklearn.linear_model import LinearRegression

otak = LinearRegression()
model = otak.fit(x_train,y_train) 
```

```python
#membuat model predict
model.predict([
    [2,1],
    [3,1],
    [1,2]
])
```

```python
y_pred = model.predict(x_test)
y_test
```

```python
y_pred
```

```python
#prediksi
y_pred = model.predict(x_test).ravel().tolist()
data = pd.DataFrame({"Nilai Sebenernya":y_test,"Prediksi":y_pred})

data.head()
```
